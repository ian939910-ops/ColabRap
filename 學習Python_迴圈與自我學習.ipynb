```
# 初始化 Initialization：設定粉紅色盒子和綠色盒子的初始值
# "<=" 右邊結果，送到左邊， 簡化為 "="
pinkBox = 0 # 粉紅色盒子一開始是 0
greenBox = 0 # 綠色盒子一開始也是 0

# 運算：改變盒子裡面的數字
pinkBox = pinkBox + 2 # 粉紅色盒子裡面的數字 加 2 = 2
greenBox = pinkBox * 2 # 綠色盒子裡面的數字，變成 粉紅色盒子(2) 乘 2 = 4

# 輸出：檢查總和並印出訊息
sum = pinkBox + greenBox # 計算兩個盒子的總和：2 + 4 = 6

# 判斷式：根據總和的大小來決定印出什麼訊息
if sum >= 7 : # 如果總和大於等於 7
  print("Sum 大於等於 7，我們下課") # 就印出這句話
else: # 否則 (如果總和小於 7)
  print("Sum 小於 7，我們繼續上課") # 就印出這句話
```

```
# ==========================================
# W3-S1 區塊 2：🏆 課堂挑戰 (CCTV 掃描線濾鏡)
# ==========================================
import cv2 # 這是用來處理圖片的函式庫，就像一個圖片編輯軟體
from google.colab.patches import cv2_imshow # 這是讓 Colab 可以顯示圖片的工具

# 1. 準備乾淨的圖片
# 從網路上下載一張圖片
!wget -q -O test.jpg https://raw.githubusercontent.com/ultralytics/yolov5/master/data/images/zidane.jpg
img = cv2.imread('test.jpg') # 讀取這張圖片到變數 img 裡面
height = img.shape[0] # 抓出圖片的總高度 (例如：720 像素)
width = img.shape[1]  # 抓出圖片的總寬度 (例如：1280 像素)

# 任務：我們希望從最上面 (Y=0) 開始，每隔 50 個像素，畫一條厚度為 2 的「綠色」水平線。
# 這樣整張圖就會佈滿橫線，看起來像監視器畫面！

# 👇 請完成下方的迴圈設定 👇
# 提示 1：y 是高度座標，範圍應該是從 0 到 圖片總高度 (height)。
# 提示 2：每隔 50 畫一條，所以「間隔」是 50。

# 這個迴圈會控制我們在圖片的哪個『高度』畫線
for y in range( 0 , height , 100 ): # 想想看，我們要從哪個高度開始 (起點)，到哪個高度結束 (終點，不包含)，每次跳多少像素 (間隔)？
    # 這行程式碼的意思：Y座標從 y 到 y+2 (代表這條線的厚度是 2 像素)，X座標 [:] 代表整條線橫跨圖片的寬度！
    # img[ Y 的範圍 , X 的範圍 ] 就是選取圖片上的某個區域
    # 等號右邊 (B, G, R) 是顏色的數值。B 是藍色，G 是綠色，R 是紅色。每個顏色從 0 到 255。
    # 👇 請把綠色的 (B, G, R) 填進去 👇
    img[ y : y+2 , : ] = ( 125 , 0 , 130 ) # 把選取的區域塗上你指定的顏色

# 顯示充滿科技感的結果！
print("\n--- 掃描線濾鏡啟動 ---")
cv2_imshow(img) # 顯示處理後的圖片
```

## 第二節 (W3-S2)：進階空間控制 (雙層迴圈與打碼) (45 分鐘)



### 👨‍🏫 教學過程細節與引導話術





* **0-15 min：降維打擊 (雙層迴圈概念)**
* **引導話術**：「如果你要在一塊空地種樹，你會怎麼種？你會先『走完第一排』，再退回原點『走第二排』。這就是迴圈包迴圈（Nested Loops）的概念。外層迴圈控制 Y (高度/排)，內層迴圈控制 X (寬度/格)！」


* **15-30 min：IndexError 陣列越界危機**
* **老師操作**：故意將迴圈範圍寫超過圖片大小，製造 `IndexError`。
* **引導話術**：「這是寫 AI 影像處理最常遇到的『致命錯誤』！圖片明明只有 720 高，你的迴圈卻叫電腦去改第 800 列的顏色，電腦就會崩潰跳出 `IndexError: index out of bounds`。看到這個錯誤，第一件事就是去檢查你的 `range` 有沒有寫超過 `shape`！」


* **30-45 min：終極 Boss 戰 (點陣網格/馬賽克雛形)**
* **引導話術**：「我們現在要在畫面上，每隔 100 像素點上一個紅色的準星點。這需要 Y 軸跟 X 軸的迴圈一起配合。請完成講義裡的雙層迴圈挑戰！」



### 💻 學生版 Colab 講義程式碼 (W3-S2)

```
# ==========================================
# W3-S2 區塊 1：🏆 雙層迴圈挑戰 (雷達定位點)
# ==========================================
# 任務：在圖片上，每隔 100 個像素的高度與寬度，畫一個 10x10 的「紅色」小方塊。
# 成果看起來會像是一張佈滿紅色定位點的網格圖！

# 重新讀取乾淨圖片 (確保每次執行都是在原圖上畫)
img_grid = cv2.imread('test.jpg')
h = img_grid.shape[0] # 圖片總高度
w = img_grid.shape[1] # 圖片總寬度

# 👇 請完成「雙層迴圈」的設定 👇
# 雙層迴圈就像在棋盤上走動：外層迴圈走『列』(Y軸)，內層迴圈走『行』(X軸)。
# 外層迴圈控制高度 (Y)，從 0 到 h，間隔 100
for y in range( 0 , h , 100 ): # 從高度 0 開始，到圖片最高點 h 為止，每隔 100 像素畫一排

    # 內層迴圈控制寬度 (X)，從 0 到 w，間隔 100
    # 注意這裡有縮排喔！代表這個迴圈是『外層迴圈裡面』的迴圈
    for x in range( 0 , h , 100 ): # 從寬度 0 開始，到圖片最寬點 w 為止，每隔 100 像素畫一個小方塊

        # 畫出 10x10 的紅色小方塊
        # 💡 防呆機制：如果 y+10 或 x+10 超過圖片邊界，會引發 IndexError！ (程式會說你超出範圍了！)
        # 但因為我們間隔是 100，所以剛好可以避開邊緣爆炸 (因為每次跳 100 像素，所以不會畫到圖片外面去)。
        # img[ Y 的起始點 : Y 的結束點 , X 的起始點 : X 的結束點 ] = 顏色
        # 👇 請填入正確的顏色數值 (B, G, R) 👇
        img_grid[ y : y+10 , x : x+10 ] = ( 10 , 10 , 100 ) # 在目前 (x, y) 的位置畫一個 10x10 的紅色方塊
print("\n--- 雷達定位點繪製完成 ---")
cv2_imshow(img_grid) # 顯示佈滿定位點的圖片
```
